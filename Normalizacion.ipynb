````
#Librerias 
import numpy as np
import pandas as pd
import cv2  
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from PIL import Image 
import matplotlib.pylab as plt
````

````
url1="https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcS9WE1IGZlR-J8_ngYfWomTeRdX17jLqmLz4Q&usqp=CAU"
img = io.imread(url1) 
cv2_imshow(img) 
norm_img = np.zeros((300, 300))
norm_img = cv2.normalize(img, norm_img, 0, 255, cv2.NORM_MINMAX)
cv2_imshow(norm_img) 
````

````
url1="https://upload.wikimedia.org/wikipedia/commons/thumb/b/b3/Colonic_Herniation_08787.jpg/290px-Colonic_Herniation_08787.jpg"
image = io.imread(url1) 
image_2 = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
cv2_imshow(image_2) 
hist = cv2.calcHist([image_2], [0], None, [256], [0, 256])
plt.plot(hist, color='gray' )
 
plt.xlabel('intensidad de iluminacion')
plt.ylabel('cantidad de pixeles')
plt.show()
````
