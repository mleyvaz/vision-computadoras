```
#Librerias 
import numpy as np
import pandas as pd
import cv2  
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from PIL import Image 
import matplotlib.pylab as plt
```

```
url2="https://upload.wikimedia.org/wikipedia/commons/thumb/f/f4/Noise_salt_and_pepper.png/220px-Noise_salt_and_pepper.png"
image1 = io.imread(url2) 
cv2_imshow(image1) 
median=cv2.medianBlur (image1, 21)
cv2_imshow(median) 
```
