# Método 2 Valor medio de lso pixeles en lo cananles 
media de lso pixels.png

```
image = imread('https://i.pinimg.com/474x/51/75/f0/5175f0051bbd40f5c72b929e53eeef7d.jpg')
imshow(image)
```

```
feature_matrix=np.zeros(image.shape)
for i in range(0,image.shape[0]):
    for j in range(0,image.shape[1]):
        feature_matrix[i][j] = ((int(image[i,j,0]) + int(image[i,j,1]) + int(image[i,j,2]))/3)
features = np.reshape(feature_matrix, (image.shape[0]*image.shape[1]*3))
features.shape, features
```
