```
#Librerias 
import numpy as np
import pandas as pd
import cv2  
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from PIL import Image 
import matplotlib.pylab as plt
```

```
#Verticales
kernel = np.array([[-1,2, -1],
                   [-1,2, -1],
                   [-1, 2,- 1]])
#Horizontales
kernel2 = np.array([[-1, -1, -1],
                   [2,2, 2],
                   [-1, -1, -1]])
#45 
kernel3 = np.array([[-1, -1, 2],
                   [-1, 2, -1],
                   [2, -1, -1]])
#135
kernel4 = np.array([[2, -1,-1],
                   [-1, 2, -1],
                   [-1, -1, 2]])
# Edge
kernel5 = np.array([[-1, -1, -1],
                   [-1, 8, -1],
                   [-1, -1, -1]])
```

```
url1="https://upload.wikimedia.org/wikipedia/commons/5/50/Vd-Orig.png"
image1 = io.imread(url1)
image = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
cv2_imshow(image)
img = cv2.filter2D(image, -1, kernel)
cv2_imshow(img)
img2 = cv2.filter2D(image, -1, kernel2)
cv2_imshow(img2)
img3 = cv2.filter2D(image, -1, kernel3)
cv2_imshow(img3)
img4 = cv2.filter2D(image, -1, kernel4)
cv2_imshow(img4)
edge = cv2.filter2D(image, -1, kernel5)
cv2_imshow(edge)
```
